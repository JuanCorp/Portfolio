```python
from difflib import SequenceMatcher
```

```python
def get_similarity_score(orig,job_name):
    score_list = list()
    for i,company in enumerate(orig):
        score_list.append(SequenceMatcher(None,company,job_name.iloc[i]).ratio())
    return score_list
```

```python
final_jobs_df['Similarity'] = get_similarity_score(final_jobs_df.Original_Company_Name,final_jobs_df.company_name)
```
